```python
import os

# PlantUML code as a string
uml_code = """
@startuml
left to right direction
actor Librarian
actor Member
rectangle "Library Management System" {
    usecase "Register Member" as UC1
    usecase "Search Book" as UC2
    usecase "Borrow Book" as UC3
    usecase "Return Book" as UC4
    usecase "Reserve Book" as UC5
    usecase "Generate Reports" as UC6
    usecase "Manage Catalog" as UC7
    usecase "Pay Fines" as UC8
}
Member -- UC1
Member -- UC2
Member -- UC3
Member -- UC4
Member -- UC5
Member -- UC8
Librarian -- UC6
Librarian -- UC7
Librarian -- UC8
UC3 ..> UC8 : "If overdue"
@enduml
"""

# Create a temporary file for the UML code
uml_file_path = "library_diagram.puml"
with open(uml_file_path, "w") as file:
    file.write(uml_code)

# Generate the PlantUML diagram as PNG
os.system(f"plantuml {uml_file_path}")

# Display the generated diagram
from IPython.display import Image
Image(filename="library_diagram.png")
```

```python
import os

# PlantUML code for the class diagram
uml_code = """
@startuml
class Member {
    +register()
    +searchBook()
    +borrowBook()
    +returnBook()
    +reserveBook()
    +payFine()
}

class Librarian {
    +addBook()
    +removeBook()
    +updateCatalog()
    +generateReports()
}

class Book {
    -title
    -author
    -ISBN
    -availability
    +getDetails()
    +updateAvailability()
}

class Transaction {
    -transactionId
    -date
    -type
    +recordTransaction()
    +getTransactionDetails()
}

class Notification {
    -message
    +sendEmail()
    +sendSMS()
}

Member --> Book : "borrows"
Member --> Transaction : "has"
Member --> Notification : "receives reminder"
Librarian --> Book : "manages"
Librarian --> Transaction : "records"
Transaction --> Notification : "sends receipt"
@enduml
"""

# Create a temporary file for the UML code
uml_file_path = "library_class_diagram.puml"
with open(uml_file_path, "w") as file:
    file.write(uml_code)

# Generate the PlantUML diagram as PNG
os.system(f"plantuml {uml_file_path}")

# Display the generated diagram
from IPython.display import Image
Image(filename="library_class_diagram.png")
```

```python
import os

# PlantUML code for the sequence diagram
uml_code = """
@startuml
actor Member
participant "Library Management System" as LMS
participant Book
participant Transaction
participant Notification
participant Librarian

Member -> LMS: searchBook()
LMS -> Book: checkAvailability()
Book -> LMS: returnAvailability()
LMS -> Member: displayResults()

Member -> LMS: borrowBook()
LMS -> Book: updateAvailability()
Book -> Transaction: createRecord()
Transaction -> LMS: recordSuccess()
LMS -> Notification: sendEmail()
Notification -> Member: borrowConfirmation()

Member -> LMS: returnBook()
LMS -> Book: updateAvailability()
Book -> Transaction: updateRecord()
Transaction -> LMS: recordReturn()
LMS -> Notification: sendReminder()
Notification -> Member: returnAcknowledgment()

Librarian -> LMS: generateReports()
LMS -> Transaction: fetchRecords()
Transaction -> LMS: returnReport()
LMS -> Librarian: displayReport()
@enduml
"""

# Create a temporary file for the UML code
uml_file_path = "library_sequence_diagram.puml"
with open(uml_file_path, "w") as file:
    file.write(uml_code)

# Generate the PlantUML diagram as PNG
os.system(f"plantuml {uml_file_path}")

# Display the generated diagram
from IPython.display import Image
Image(filename="library_sequence_diagram.png")
```

```python
import os

# PlantUML code for the activity diagram
uml_code = """
@startuml
start
:User Accesses Library System;
:Login with Credentials;
if (Are Credentials Valid?) then (Yes)
    :Show User Dashboard;
    repeat
        :Select an Action;
        switch (Chosen Action)
            case (Search Book)
                :Enter Search Criteria;
                :Display Search Results;
            case (Borrow Book)
                :Check Book Availability;
                if (Is Available?) then (Yes)
                    :Issue Book;
                    :Update Inventory;
                    :Send Confirmation Email;
                else (No)
                    :Show Unavailable Message;
                endif
            case (Return Book)
                :Update Book Status;
                :Record Return Transaction;
                :Send Return Confirmation;
            case (Pay Fine)
                :Process Fine Payment;
                :Update Payment Records;
                :Generate Receipt;
        endswitch
    repeat while (User Continues?)
    :Logout;
    :End Session;
else (No)
    :Show Login Error;
endif
stop
@enduml
"""

# Create a temporary file for the UML code
uml_file_path = "library_activity_diagram.puml"
with open(uml_file_path, "w") as file:
    file.write(uml_code)

# Generate the PlantUML diagram as PNG
os.system(f"plantuml {uml_file_path}")

# Display the generated diagram
from IPython.display import Image
Image(filename="library_activity_diagram.png")
```

```python
import os

# PlantUML code for the component diagram
uml_code = """
@startuml
component "User Interface" as UI
component "Authentication Service" as Auth
component "Library Services" as Library
component "Book Catalog" as Catalog
component "Transaction Service" as Transaction
component "Notification Service" as Notify
component "Database" as DB

UI --> Auth : User Login
Auth --> DB : Verify Credentials
Auth --> UI : Authentication Success/Failure
UI --> Library : Request Book Info
Library --> Catalog : Fetch Book Details
Catalog --> DB : Retrieve Data
Library --> UI : Display Book Details
UI --> Transaction : Borrow/Return Book
Transaction --> Library : Validate Request
Transaction --> DB : Update Records
Transaction --> Notify : Send Transaction Alert
Transaction --> UI : Show Transaction Status
UI --> Notify : Overdue Reminders
Notify --> UI : Display Notifications
@enduml
"""

# Create a temporary file for the UML code
uml_file_path = "library_component_diagram.puml"
with open(uml_file_path, "w") as file:
    file.write(uml_code)

# Generate the PlantUML diagram as PNG
os.system(f"plantuml {uml_file_path}")

# Display the generated diagram
from IPython.display import Image
Image(filename="library_component_diagram.png")
```

```python
import os

# PlantUML code for the component diagram with nodes
uml_code = """
@startuml
node "User Device" {
    component "Web Browser" as Browser
}

node "Library Server" {
    component "Web Server" as WebServer
    component "Application Server" as AppServer
    database "Library Database" as DB
    component "Notification Service" as Notify
}

Browser --> WebServer : HTTP Request (Search, Borrow, Return)
WebServer --> AppServer : Process Request
AppServer --> DB : Fetch/Update Data
DB --> AppServer : Send Response
AppServer --> Notify : Send Email/SMS Alerts
AppServer --> WebServer : Send Processed Data
WebServer --> Browser : Display Result
@enduml
"""

# Create a temporary file for the UML code
uml_file_path = "library_component_with_nodes_diagram.puml"
with open(uml_file_path, "w") as file:
    file.write(uml_code)

# Generate the PlantUML diagram as PNG
os.system(f"plantuml {uml_file_path}")

# Display the generated diagram
from IPython.display import Image
Image(filename="library_component_with_nodes_diagram.png")
```

```python
import os

# PlantUML code for the state diagram
uml_code = """
@startuml
[*] --> Idle
Idle --> LoggingIn : User Enters Credentials
LoggingIn --> LoggedIn : Authentication Successful
LoggingIn --> Blocked : Too Many Failed Attempts
Blocked --> Idle : Reset Password
LoggedIn --> SearchingBook : Search for a Book
SearchingBook --> LoggedIn : Book Found / Not Found
LoggedIn --> BorrowingBook : Borrow a Book
BorrowingBook --> BookIssued : Borrowing Successful
BorrowingBook --> LoggedIn : Borrowing Failed
BookIssued --> LoggedIn
LoggedIn --> ReturningBook : Return a Book
ReturningBook --> BookReturned : Return Successful
ReturningBook --> LoggedIn : Return Failed
BookReturned --> LoggedIn
LoggedIn --> LoggingOut : Logout Request
LoggingOut --> Idle : Session Ended
@enduml
"""

# Create a temporary file for the UML code
uml_file_path = "library_state_diagram.puml"
with open(uml_file_path, "w") as file:
    file.write(uml_code)

# Generate the PlantUML diagram as PNG
os.system(f"plantuml {uml_file_path}")

# Display the generated diagram
from IPython.display import Image
Image(filename="library_state_diagram.png")
```

```python
import os

# PlantUML code for the collaboration diagram
uml_code = """
@startuml
title Collaboration Diagram for Library Management System

object User
object LibrarySystem
object BookCatalog
object Transaction
object Notification

' User login process
User -down-> LibrarySystem : (1) login()
LibrarySystem -down-> User : (2) authenticationSuccess()
LibrarySystem -down-> User : (3) authenticationFailed()
User -down-> LibrarySystem : (4) resetPassword()

' Searching for books
User -down-> LibrarySystem : (5) searchBook()
LibrarySystem -down-> BookCatalog : (6) findBook()
BookCatalog -down-> LibrarySystem : (7) returnBookDetails()
LibrarySystem -down-> User : (8) displayBookInfo()

' Borrowing books
User -down-> LibrarySystem : (9) borrowBook()
LibrarySystem -down-> Transaction : (10) createTransaction()
Transaction -down-> LibrarySystem : (11) confirmBorrowing()
LibrarySystem -down-> Notification : (12) sendEmail()
Notification -down-> User : (13) borrowConfirmation()

' Returning books
User -down-> LibrarySystem : (14) returnBook()
LibrarySystem -down-> Transaction : (15) updateTransaction()
Transaction -down-> LibrarySystem : (16) confirmReturn()
LibrarySystem -down-> Notification : (17) sendReturnConfirmation()
Notification -down-> User : (18) returnReceipt()

' Logout process
User -down-> LibrarySystem : (19) logout()
LibrarySystem -down-> User : (20) logoutSuccess()

@enduml
"""

# Create a temporary file for the UML code
uml_file_path = "library_collaboration_diagram.puml"
with open(uml_file_path, "w") as file:
    file.write(uml_code)

# Generate the PlantUML diagram as PNG
os.system(f"plantuml {uml_file_path}")

# Display the generated diagram
from IPython.display import Image
Image(filename="library_collaboration_diagram.png")
```
