```python
# Install PlantUML (if not already installed)
!apt-get install -y plantuml

# PlantUML code for the activity diagram
uml_code = """
@startuml
start
:User Opens E-commerce Website;
:Enter Credentials;
if (Are Credentials Valid?) then (Yes)
    :Display Homepage;
    repeat
        :Choose an Action;
        switch (Selected Action)
            case (Browse Products)
                :Retrieve Product List;
                :Display Products;
            case (Add to Cart)
                :Select Product;
                if (Is Product in Stock?) then (Yes)
                    :Add Product to Cart;
                    :Update Cart;
                else (No)
                    :Show Out of Stock Message;
                endif
            case (Place Order)
                :Review Cart;
                :Proceed to Checkout;
                :Enter Payment Details;
                if (Is Payment Successful?) then (Yes)
                    :Confirm Order;
                    :Send Order Confirmation Email;
                else (No)
                    :Show Payment Failure Message;
                endif
            case (Track Order)
                :Enter Order ID;
                :Retrieve Order Status;
                :Display Order Status;
            case (Reset Password)
                :User Requests Password Reset;
                :Send Reset Email;
                :User Enters New Password;
                :Password Updated;
        endswitch
    repeat while (User Wants Another Action?)

    :Logout;
    :End Session;
else (No)
    :Show Login Error;
endif
stop
@enduml
"""

# Create a temporary file for the UML code
uml_file_path = "/content/activity_diagram.puml"
with open(uml_file_path, "w") as file:
    file.write(uml_code)

# Generate the PlantUML diagram as PNG
import os
os.system(f"plantuml {uml_file_path}")

# Display the generated diagram
from IPython.display import Image
Image(filename="/content/activity_diagram.png")
```
